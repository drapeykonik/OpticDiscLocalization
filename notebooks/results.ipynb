# Results

```python
import json
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
sns.set_theme(style="darkgrid", font_scale=1.4)
```

```python
exp = 12
```

```python
batch_loss = dict()
batch_loss["train"] = np.array(json.load(open(f"results/run-exp{exp}_logs-tag-Loss (batch)_train.json")))
batch_loss["valid"] = np.array(json.load(open(f"results/run-exp{exp}_logs-tag-Loss (batch)_valid.json")))
```

```python
fig = plt.figure(figsize=(14, 9))
plt.plot(batch_loss["train"][:, 1], batch_loss["train"][:, 2])
plt.title("VGG model training. Train loss by batch")
plt.xlabel("Batch number")
plt.ylabel("Loss value")
```

```python
fig = plt.figure(figsize=(14, 9))
plt.plot(batch_loss["valid"][:, 1], batch_loss["valid"][:, 2])
plt.title("VGG model training. Valid loss by batch")
plt.xlabel("Batch number")
plt.ylabel("Loss value")
```

```python
epoch_loss = dict()
epoch_loss["train"] = np.array(np.array(json.load(open(f"results/run-exp{exp}_logs-tag-Loss (epoch)_train.json"))))
epoch_loss["valid"] = np.array(np.array(json.load(open(f"results/run-exp{exp}_logs-tag-Loss (epoch)_valid.json"))))
```

```python
fig = plt.figure(figsize=(12, 8))
plt.plot(epoch_loss["train"][:, 1], epoch_loss["train"][:, 2], label="Training")
plt.plot(epoch_loss["valid"][:, 1], epoch_loss["valid"][:, 2], label="Validation")
plt.title("VGG model training. Loss by epoch")
plt.xlabel("Epoch")
plt.ylabel("Loss value")
plt.legend()
```

```python
loc_acc = dict()
loc_acc["r = 25"] = np.array(json.load(open(f"results/run-exp{exp}_logs-tag-loc_acc_r25.json")))
loc_acc["r = 50"] = np.array(json.load(open(f"results/run-exp{exp}_logs-tag-loc_acc_r50.json")))
loc_acc["r = 75"] = np.array(json.load(open(f"results/run-exp{exp}_logs-tag-loc_acc_r75.json")))
loc_acc["r = 100"] = np.array(json.load(open(f"results/run-exp{exp}_logs-tag-loc_acc_r100.json")))
```

```python
fig = plt.figure(figsize=(14, 9))
for tag, values in loc_acc.items():
    plt.plot(values[:, 1], values[:, 2], label=tag)
plt.title("Localization Accuracy on test by epoch")
plt.xlabel("Epoch")
plt.ylabel("Metric value")
plt.legend()
```
